```python
rotg=pd.read_csv('data/20160312_RunninoftheGreen4m.csv')
```

```python
rotg.head()
```

```python
temp2=pd.DataFrame()
#temp2['PLACE']=results.Place
temp2['id']=rotg.index
temp2['time']=rotg.Time.apply(lambda x: '00:'+x.split(':')[0]+':'+x.split(':')[1])
temp2['race_id']=3
temp2['runner_id']=0
temp2['first_name']=rotg.Name.apply(lambda x: x.split(' ',1)[0])
temp2['last_name']=rotg.Name.apply(lambda x: x.split(' ',1)[1])
temp2['gender']=rotg.Sex
temp2['city']=rotg.City
temp2['state']=rotg.State
temp2['age']=rotg.Age
temp2['bib']=rotg.Bib
temp2['pace']=rotg.Pace
temp2['splits']='00:00'
#temp2['YEAR']=2016
```

```python
temp2.head()
```

```python
temp2.to_sql('results_result',conn, if_exists='replace',index=False)
#help(temp2.to_sql)
```
