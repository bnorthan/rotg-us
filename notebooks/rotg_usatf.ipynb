```python
import pandas as pd
```

```python
rotg=pd.read_csv('../data/rotg-m9.csv')
usatf=pd.read_excel('../data/usatf.xlsx')
clubs=pd.read_csv('../data/clubs.csv')
```

```python
usatf=pd.merge(usatf, clubs, how='left', left_on='club', right_on='clubid')
```

```python
rotg=pd.merge(rotg, usatf2, how='left', left_on='usatf_num', right_on='num')
```
