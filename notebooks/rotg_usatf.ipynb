```python
import pandas as pd
```

```python
rotg=pd.read_csv('../data/rotg-clubs.csv')
usatf=pd.read_excel('../data/usatf2.xlsx')
clubs=pd.read_csv('../data/clubs.csv')
```

```python
usatf=pd.merge(usatf, clubs, how='left', left_on='Club Affiliation', right_on='clubid')
usatf.head()
```

```python
usatf2=usatf[['Membership #', 'clubname','clubid']]
```

```python
rotg=pd.merge(rotg, usatf2, how='left', left_on='usatfnum', right_on='Membership #')
```

```python
rotg['Membership #']=rotg['Membership #'].fillna("")
```
